```python
df = data_clening.clean_data()

plt.figure(figsize=(10,5))
sns.countplot(x='ジャンル', data=df)
plt.xticks(rotation=45, fontsize=10)
plt.show()

# 各ジャンルの平均価格
average_price_per_genre = df.groupby('ジャンル')['価格'].mean().reset_index()
sns.barplot(x='価格', y='ジャンル', data=average_price_per_genre)
plt.show()


# 発売年ごとの商品数を表示
df = data_clening.clean_data()
plt.figure(figsize=(10,5))
# 年ごとの商品数
df['発売年'] = df['発売日'].dt.year
count_per_year = df['発売年'].value_counts().sort_index()
sns.lineplot(x=count_per_year.index, y=count_per_year.values)
plt.xticks(rotation=45, fontsize=9)
x_pos = 2018
y_pos = count_per_year.max() + 5  # 最大値から少し上にオフセット
plt.text(x=x_pos, y=y_pos, s="Switchバードとamiiboが多い", fontsize=12, color='red')
plt.show()

# # 商品の状態による価格の分布をヒストグラムで表示
# plt.figure(figsize=(10,6))
# sns.histplot(data=df, x='価格', hue='状態')
# plt.show()

# 時間経過による価格変動
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='発売年', y='価格')

# X軸の位置はデータセット内の最新の年（あるいは任意の年）に設定
x_pos = df['発売年'].max()

# Y軸の位置は価格の最大値よりも少し高く設定
y_pos = df['価格'].max() + (df['価格'].max() * 0.01)

# 注釈を加える
plt.text(x=x_pos, y=y_pos, s="Switchバードが平均価格を上げている", fontsize=12, color='red', horizontalalignment='right')

plt.show()
```

```python
import pandas as pd
# '発売日'列をdatetime型に変換し、エラーがある場合はNaTに置き換え
df['発売日'] = pd.to_datetime(df['発売日'], errors='coerce')

# '発売日'から年を抽出
df['発売年'] = df['発売日'].dt.year

# 2016年から2020年までの商品のデータを抽出
filtered_df = df[df['発売年'].between(2010, 2020)]

# ジャンルごとに商品数をカウント
genre_counts = filtered_df['ジャンル'].value_counts()

# ジャンルでフィルタリングして、商品数が10個以上のものだけを取得
filtered_genres = genre_counts[genre_counts >= 10].index.tolist()

# 条件に合うジャンルの商品だけを表示
df_filtered_final = filtered_df[filtered_df['ジャンル'].isin(filtered_genres)]

# 結果のデータフレームを表示
display(df_filtered_final)
```
